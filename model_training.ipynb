# Dimension reductions

```python
pca_threshold = 0.95
# Dimension reduction
pca = PCA()
pca.fit(num_data)

# Compute components
reduced_data = PCA(n_components=pca_threshold).fit_transform(
    num_data[feature_imp["Feature"].loc[feature_imp["gain"]>10].values])
```

```python
reduced_data.plot()#.shape
```

```python
# Set parameters used for dimension reduction
x = 10
n_neighbors_articles = x * 2
local_connectivity_articles = x
#min_dist_articles = np.round(
#    np.abs(num_data.median(axis=1).median() - num_data.mean(axis=1).median()),
#    5)
min_dist_articles = 0.1
# Run dimension reduction
reducer = umap.UMAP(n_neighbors=n_neighbors_articles,
                    local_connectivity=local_connectivity_articles,
                    min_dist=min_dist_articles,
                    n_components=10,
                    random_state=42,
                    n_epochs=500,
                    metric="cosine")

UMAP = reducer.fit_transform(num_data.iloc[:,:-4])


```

```python
# Extract identifers 
processed_data = combined_data[[
    'id', 'date', 'league_id', 'league_name', 'home_team_id', 'home_team_name',
    'away_team_id', 'away_team_name', "result"
]]

# Add in umaps 
processed_data['umap_projection_0'] = UMAP[:, 0]
processed_data['umap_projection_1'] = UMAP[:, 1]
processed_data['umap_projection_2'] = UMAP[:, 2]
processed_data['umap_projection_3'] = UMAP[:, 3]
processed_data['umap_projection_4'] = UMAP[:, 4]
processed_data['umap_projection_5'] = UMAP[:, 5]
processed_data['umap_projection_6'] = UMAP[:, 6]
processed_data['umap_projection_7'] = UMAP[:, 7]
processed_data['umap_projection_8'] = UMAP[:, 8]
processed_data['umap_projection_9'] = UMAP[:, 9]

# Turn pca data to dataframe and fix col names
pca_cols = pd.DataFrame(reduced_data)
pca_cols.columns = [str(x) + "_pca" for x in pca_cols.columns]
processed_data = pd.concat([processed_data, pca_cols], axis=1)
```

```python
processed_data["target"] = np.nan
processed_data["target"].loc[processed_data["result"] == "1"] = 0
processed_data["target"].loc[processed_data["result"] == "x"] = 1
processed_data["target"].loc[processed_data["result"] == "2"] = 2

train_data = processed_data[processed_data.date <"2021-03-04"]
vaild_data = processed_data[(processed_data.date >= "2021-03-04") & (processed_data.date <= "2021-05-01")]
test_data = processed_data[processed_data.date > "2021-05-01"]
```
